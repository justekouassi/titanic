```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
m = 100
X = np.linspace(0, 10, m).reshape(m, 1)
y = X + np.random.randn(m, 1)
```

```python
plt.scatter(X, y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
```

```python
modele_lineaire = LinearRegression()
modele_lineaire.fit(X, y) # entraînement du modèle
modele_lineaire.score(X, y) # coefficient de détermination
```

```python
predictions = modele_lineaire.predict(X)
plt.scatter(X, y)
plt.plot(X, predictions, c='red')
plt.show()
```

```python
y = X**2 + np.random.randn(m, 1)
plt.scatter(X, y)
plt.show()
```

```python
modele_carree = SVR(C=100)
modele_carree.fit(X, y) # entraînement du modèle
modele_carree.score(X, y) # coefficient de détermination
```

```python
predictions = modele_carree.predict(X)
plt.scatter(X, y)
plt.plot(X, predictions, c='r')
plt.show()
```
